```python
import matplotlib.pyplot as pyplot
import numpy
```

# Binomial Distribution

```python
#Plays the ball drawing game
def ballDrawingGame(n, p):
    success = 0
    for i in range(0,n):
        probability = numpy.float32(numpy.random.random())
        if (probability <= p):
            success = success + 1
    return success
```

```python
#Plays multiple ball drawing games
def ballDrawingGames(trials, n, p):
    successes = []
    for i in range(0, trials):
        successes.append(ballDrawingGame(n, p))
    return successes
```

```python
#Shows the ball drawing game results
results = ballDrawingGames(100000, 50, 0.7)
pyplot.hist(results)
pyplot.show()
```
